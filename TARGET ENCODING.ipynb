```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'x_0': ['dog'] * 5 + ['cat'] * 5,
    'x_1': ['wolf'] * 9 + ['tiger'] * 1,
    'y': [1, 1, 1, 1, 0, 1, 0, 0, 0, 0]
})
```

```python
m1=df.groupby('x_0')['y'].mean()
m2= df.groupby('x_1')['y'].mean()
```

```python
df['x_0'] = df['x_0'].map(m1)
```

```python
df['x_1'] = df['x_1'].map(m2)
```

```python
def calc_smooth_mean(df, by, on, m):
    # Compute the global mean
    mean = df[on].mean()

    # Compute the number of values and the mean of each group
    agg = df.groupby(by)[on].agg(['count', 'mean'])
    counts = agg['count']
    means = agg['mean']

    # Compute the "smoothed" means
    smooth = (counts * means + m * mean) / (counts + m)

    # Replace each value by the according smoothe
    return df[by].map(smooth)
```

```python

df['x_0'] = calc_smooth_mean(df, by='x_0', on='y', m=10)
df['x_1'] = calc_smooth_mean(df, by='x_1', on='y', m=10)
```
